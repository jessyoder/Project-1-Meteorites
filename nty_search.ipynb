```python
# Dependencies
import requests
import pandas as pd
from pprint import pprint
from config import nyt_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
# Search for articles that mention meteorites
query = "meteorite"
```

```python
# Build query URL
query_url = url + "api-key=" + nyt_key + "&q=" + query
```

```python
# Request articles
articles = requests.get(query_url).json()

# The "response" property in articles contains the actual articles
# list comprehension.
articles_list = articles['response']['docs']
pprint(articles_list)
```

```python
abstract_list = []
headline_list = []
lead_par_list = []
pub_date = []
art_snippet = []
url = []

for articles in articles_list:
    abstract = articles['abstract']
    abstract_list.append(abstract)
    headline = articles['headline']['main']
    headline_list.append(headline)
    lead_paragraph = articles['lead_paragraph']
    lead_par_list.append(lead_paragraph)
    publication_date = articles['pub_date']
    pub_date.append(publication_date)
    snippet = articles['snippet']
    art_snippet.append(snippet)
    nyt_site = articles['web_url']
    url.append(nyt_site)
    
nyt_meteor_df = pd.DataFrame({'Abstract': abstract_list,
                        'Headline': headline_list,
                        'Lead Paragraph': lead_par_list,
                        'Publication Date': pub_date,
                        'Article Snippet': art_snippet,
                        'URL': url})
nyt_meteor_df
```

```python
nyt_meteor_df.to_csv("nyt_article_info", index=False, header=True)
```
